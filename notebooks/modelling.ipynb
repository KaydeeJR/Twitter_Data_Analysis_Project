# Topic modelling<a class="anchor" id="modelling"></a>

## Dictionary corpus<a class="anchor" id="corpus"></a>

```python
topic_model = TopicModelling()
tweet_mappings = topic_model.make_dictionary(words_list)
```

```python
parent_dir = os.path.dirname(os.getcwd())
topic_model.save_dictionary(tweet_mappings,'global_mappings')
```

## Bag of words model<a class="anchor" id="bow"></a>

```python
bow = topic_model.create_bow(words_list, tweet_mappings)
topic_model.serialize_bow('global_bow',bow)
```

## LDA model<a class="anchor" id="lda"></a>

**Reading from files in current directory in order to avoid recreating the mappings and bow**

```python
saved_bow = topic_model.load_bow('global_bow')
saved_mapping = topic_model.load_dictionary('global_mappings')
```

```python
lda = topic_model.create_lda_model(saved_bow, saved_mapping)
```

**Creating the LDA model takes approximately 50s**

Saving the LDA model in order to avoid recreating it and save substantial computing time

```python
topic_model.save_lda_model(lda, 'global_lda')
```

# Visualizing results<a class="anchor" id="visualizing"></a>

```python
lda_model = topic_model.load_lda_model('global_lda')
```

## Topic visualization using pyLDAvis<a class="anchor" id="topics"></a>

The pyLDAvis package produces an interactive plot where each circle is a topic and the size of the circle represents the abundance of that topic in the corpus.

Along with each topic are the associated words that go with it.

```python
pyLDAvis.enable_notebook()
lda_visualisation = topic_model.visualize_lda_results(lda_model, saved_bow, saved_mapping)
display(lda_visualisation)
```
